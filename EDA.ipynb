## Imports and Set Up

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
pd.set_option('display.max_columns', None)
```

## Load Dataset

```python
df_orig = pd.read_csv('./creditcard_2023.csv', index_col=0)
df_train, df_test = train_test_split(df_orig, test_size=0.1)
display(df_train)
display(df_train.describe())
```

```python
# Total number of empty values
df_orig.isnull().sum().sum()
```

> **Notes**
>
> - Features V1 - V28 are rescaled with `0` mean and `1` std.
> - Classes are **balanced**.
> - There are no missing values.

## `Amount` analysis

```python
sns.histplot(df_train, x='Amount', hue='Class', multiple='stack')
plt.show()
```

```python
sns.boxplot(df_train, x='Class', y='Amount')
```

```python
df_train[['Amount', 'Class']].corr(method='spearman')
```

> **Notes**
>
> - `Amount` has **uniform** distribution. 
> - `Amount` has **no correlation** with the target variable

## Explore other features (Vn)

```python
v_cols = [c for c in df_train.columns if c[0] == 'V']
v_corr = pd.concat([df_train[v_cols], df_train['Class']], axis=1).corr(method='spearman')['Class']\
    .sort_values(ascending=False, key=abs).drop('Class')

# plot
sns.barplot(x=v_corr, y=v_corr.index)
plt.grid(True, axis='x')
plt.title('Spearman correlation of features with the target')
plt.show()
```

```python
# Hist plots of anonymous values
df_melted = df_train[v_cols].melt(var_name='column')
sns.FacetGrid(df_melted, col='column', col_wrap=4, sharey=False, sharex=False).map(plt.hist, 'value', bins=20)
plt.show()
```
